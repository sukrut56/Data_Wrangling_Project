```python
import matplotlib.pyplot as plt

# Create a dictionary to hold the counts of zeros for each column
counts = {}
for column_name in df1.columns:
    column = df1[column_name]
    count = (column == 'Unknown').sum()
    counts[column_name] = count

# Convert the dictionary to a pandas Series and plot a bar graph
plt.figure(figsize=(15, 10), dpi=80)
counts_series = pd.Series(counts)
counts_series.plot(kind='bar')
plt.xlabel('Column')
plt.ylabel('Unknowns')
plt.title('Count of Unknown in each column')
plt.show()
```

```python
import matplotlib.pyplot as plt
null = df1.isnull().sum()
plt.figure(figsize=(15, 10), dpi=80)
null.plot(kind='bar')
plt.title('Null Values in DataFrame')
plt.xlabel('Columns')
plt.ylabel('Number of Null Values')
plt.show()
```
